### 게임 평점 시각화

```python
user_mean_list = []
critic_mean_list = []

for row in tqdm(best_game_df.itertuples(), 'list building...', best_game_df.shape[0]):
    user_mean_list.append(row[4]['user']['mean'])
    critic_mean_list.append(row[4]['critic']['mean'])

user_mean_list
```

```python
# 1. 스타일 설정
plt.style.use('default')
plt.rcParams['figure.figsize'] = (6, 6)
plt.rcParams['font.size'] = 12

# 2. 데이터 준비
x = np.arange(150)
y1 = np.array(user_mean_list)
y2 = np.array(critic_mean_list)

# 3. 그래프 그리기
fig, ax1 = plt.subplots()

ax1.plot(x, y1, color='#33FF88', markersize=7, linewidth=3, alpha=0.7, label='User')
ax1.set_ylim(8.9, 9.8)
ax1.set_ylabel('user score')
ax1.tick_params(axis='both', direction='in')

ax2 = ax1.twinx()
ax2.plot(x, y2, color='#FF6666', markersize=7, linewidth=3, alpha=0.7, label='Critic')
ax2.set_ylim(40, 100)
ax2.set_ylabel('critic score')
ax2.tick_params(axis='y', direction='in')

ax1.legend(loc='upper left')
ax2.legend(loc='upper right')
plt.show()
```

### 게임 플랫폼(기기) 분포 시각화

```python
pc_or_not_list = [0, 0]

for row in tqdm(best_game_df.itertuples(), 'list building...', best_game_df.shape[0]):
    pc_or_not_list[0 if row[1] == 'PC' else 1] += 1

pc_or_not_list
```

```python
# 1. 스타일 설정
plt.style.use('default')
plt.rcParams['figure.figsize'] = (6, 6)
plt.rcParams['font.size'] = 12

# 2. 데이터 준비
x = np.array(['PC', 'Console'])
y = np.array(pc_or_not_list)

# 3. 그래프 그리기
plt.barh(x, y)
plt.show()
```

```python
platform_value_count = best_game_df['platform'].value_counts().copy()
```

```python
# 1. 스타일 설정
plt.style.use('default')
plt.rcParams['figure.figsize'] = (6, 6)
plt.rcParams['font.size'] = 12

# 2. 데이터 준비
x = np.array(platform_value_count.index.copy())
y = platform_value_count.values

# 3. 그래프 그리기
plt.barh(x, y)
# plt.xticks(rotation=55)
plt.show()
```
